```python
base = base.drop(columns=["song_name"])
```

```python
#2 - Faça a dummificação nas variáveis categóricas.
categorica = pd.get_dummies(base[["audio_mode",]])
```

```python
categorica
```

```python
base2 = base.drop(columns=["audio_mode",])
base_nova = pd.concat([categorica, base2], axis=1)
```

```python
#3 - Separes as bases em variáveis independentes e dependentes.
x = base_nova.drop(columns=["audio_valence"])
y = base_nova["audio_valence"]
```

```python
x
```

```python
#4 - Crie a base de treino e de teste.
from statistics import linear_regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.3,random_state=28 )

```

```python
#5 - Crie o classificador


modelo = LinearRegression()
modelo.fit(xtrain, ytrain)
```

```python
modelo.acuracia = modelo.score(xtest, ytest)
modelo.acuracia
```
