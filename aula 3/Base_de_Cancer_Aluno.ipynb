```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
base = pd.read_csv('breast-cancer-dataset.csv')
```

```python
base
```

```python
#1 - Apague a coluna S/N
base.drop(columns=["S/N"], inplace=True)
#2 - Faça a dummificação nas variáveis categóricas.
conv = {
    "Benign": 0,
    "Malignant": 1
}
type(base["Metastasis"][0])


```

```python
base = base[(base["Metastasis"] == "0") | (base["Metastasis"] == "1")].copy()

base["Metastasis"] = pd.to_numeric(base["Metastasis"])
    # base.loc[base["Metastasis"] != "0" and base["Metastasis"] != "1"]
    
    # convert = int(item)

    # item = convert
    
```

```python
base["Diagnosis Result"] = base["Diagnosis Result"].map(conv)
base.drop(columns=["Breast"], inplace=True)

categories = base[["Menopause", "Metastasis","Breast Quadrant", "History"]]
categorica = pd.get_dummies(categories)  
```

```python
#3 - Separes as bases em variáveis independentes e dependentes.
base.drop(columns=["Menopause", "Metastasis", "History","Breast Quadrant"], inplace= True)
base

```

```python
categorica
```

```python
base
```

```python
base["Tumor Size (cm)"] = pd.to_numeric(base["Tumor Size (cm)"])
base["Inv-Nodes"] = pd.to_numeric(base["Inv-Nodes"])
base["Year"] = pd.to_numeric(base["Year"])
```

```python
type(base["Year"][6])
```

```python
nbase = pd.concat([base, categorica], axis=1)
```

```python
nbase
```

```python
#determi8nando as bases de treino x e y

y = nbase["Diagnosis Result"]
x = nbase.drop(columns=["Diagnosis Result"])

```

```python
#4 - Crie a base de treino e de teste.
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.3, random_state=28)

# print("Número de amostras em xtrain:", len(xtrain))
# print("Número de amostras em ytrain:", len(ytrain))
#5 - Crie o classificador
modelo = DecisionTreeClassifier()

modelo.fit(xtrain, ytrain)



predito = modelo.predict(xtest)
# predito = pd.Series(predito, name="valor predito")
# real = pd.Series(ytest, name = "valor real")

# modelo_resultados = pd.concat([predito, real], axis=1)

# modelo_resultados


#6 - Avalie o classificador
acuracia = (ytest == predito).mean()
print(f"O modelo acertou {acuracia*100}% das previsões")


```
