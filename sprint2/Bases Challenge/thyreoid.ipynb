```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
base_tyreoid = pd.read_csv("thyroidDF.csv")
```

```python
base_tyreoid.head(50)
base_tyreoid = base_tyreoid.drop(columns=["T4U", "TBG","patient_id"])
```

```python
base_tyreoid = base_tyreoid.dropna()
```

```python
base_tyreoid["target"].value_counts
```

```python
categorica = pd.get_dummies(base_tyreoid, columns=[
    "age",
    "sex",
    "on_thyroxine",
    "query_on_thyroxine",
    "on_antithyroid_meds",
    "sick",
    "pregnant",
    "thyroid_surgery",
    "I131_treatment",
    "query_hypothyroid",
    "query_hyperthyroid",
    "lithium",
    "goitre",
    "hypopituitary",
    "psych",
    "TSH_measured",
    "T3_measured",
    "TT4_measured",
    "T4U_measured",
    "FTI_measured",
    "TBG_measured",
    "referral_source",
    "target"
])
```

```python
pd.set_option("display.max_columns", None)

categorica.head(5)
```

```python
x = categorica.drop(columns=["tumor"])
y = categorica["tumor"]
```

```python
xtrain, ytrain, xtest, ytest = train_test_split(x, y , test_size=0.3, random_state=28)
```

```python
modelo = DecisionTreeClassifier()

modelo.fit(xtrain, ytrain)

modelo.acuracia = modelo.score(xtest, ytest)
```
