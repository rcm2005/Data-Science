### Dummy, função da biblioteca pandas que realiza o one hot encoding

`````python
base_categorica = pd.get_dummies(base_cardiac2, columns=["Year of Hospital Discharge","Procedure"],dtype=int)
`````

`````python
base_categorica #verifico como ficou a base apos o dummy
`````

## Preparação para o modelo de Arvore De Decisão Classificatoria

`````python
X = base_categorica.drop(columns=["Comparison Results"])    #todas as colunas da base, tirando a coluna alvo
y = base_categorica["Comparison Results"]                   # coluna alvo separada da base
`````

### Preparação da base de treino e teste para o modelo
usamos 30% da base p/ teste e pegamos uma seed (24) para o valor ser exatamente o mesmo independente da maquina

`````python
xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size=0.3, random_state=24)
`````

### Variavel que recebe o modelo de decisao
O metodo 
```python
    modelo.fit(xtrain, ytrain)
```` 
fica responsavel por pegar a base de treino e aplicar ao modelo de classsificação

`````python
modelo = DecisionTreeClassifier()

modelo.fit(xtrain, ytrain)
`````

##### testa a acuracia do modelo de classificação do modelo

`````python
modelo.acuracia = modelo.score(xtest, ytest)
`````
