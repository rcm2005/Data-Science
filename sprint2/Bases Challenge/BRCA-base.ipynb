```python
base_BRCA = base_BRCA.drop(columns=["Patient_ID","Date_of_Surgery","Date_of_Last_Visit"])
```

```python
#categorica ["Gender", "Histology", "Patient_Status" ,"Surgery_type","Date_of_Surgery","ER status","PR status","HER2 status"]
```

```python
categorica = pd.get_dummies(base_BRCA, columns=["Gender","Histology", "Patient_Status" ,"Surgery_type","ER status","PR status","HER2 status"], dtype=int)
```

```python
x = categorica.drop(columns=["Tumour_Stage"])
y = categorica["Tumour_Stage"]
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.3, random_state=28)
```

```python
modelo = DecisionTreeClassifier()
modelo.fit(xtrain, ytrain)

modelo.acuracia = modelo.score(xtest, ytest)
modelo.acuracia
```
