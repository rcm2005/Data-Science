```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
base_thoracic = pd.read_csv("ThoraricSurgery.csv")
```

```python
#["id"]
base_thoracic = base_thoracic.drop(columns=["id"])
```

```python
#['DGN', 'PRE4', 'PRE5', 'PRE6', 'PRE7', 'PRE8', 'PRE9', 'PRE10','PRE11', 'PRE14', 'PRE17', 'PRE19', 'PRE25', 'PRE30', 'PRE32']

categorica = pd.get_dummies(base_thoracic, columns=['DGN', 'PRE4', 'PRE5', 'PRE6', 'PRE7', 'PRE8', 'PRE9', 'PRE10','PRE11', 'PRE14', 'PRE17', 'PRE19', 'PRE25', 'PRE30', 'PRE32'])
```

```python
categorica
```

```python
y = categorica["Risk1Yr"]
x = categorica.drop(columns=["Risk1Yr"])
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.3, random_state=28)
```

```python
modelo = DecisionTreeClassifier()

modelo.fit(xtrain,ytrain)
```

```python
modelo.acuracia = modelo.score(xtest,ytest)
modelo.acuracia
```
