```python
import pandas as pd
```

# Definindo dataset

```python
base = pd.read_csv('bank-full.csv', delimiter=';')
```

```python
base
```

```python
base.head()
```

## Realizando dummificação das variáveis categoricas

```python
categorica = pd.get_dummies(base, columns=["job","marital", "education","default", "housing", "loan", "contact", "month", "poutcome"], dtype=int)
```

### Retirando variaveis categóricas não dummificadas

```python
base_drop = base.drop(columns=["job","marital", "education","default", "housing", "loan", "contact", "month", "poutcome"], inplace=True)
```

#### Novo dataset

```python
base_nova = pd.concat([categorica,base_drop], axis=1)

base_nova
```

#### Importando sklearn

```python
#TTS e RandomForest

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

```

```python
#2 - Separes as bases em variáveis independentes e dependentes.

x= base_nova.drop(columns= "y")
y = base_nova["y"]
```

```python
#3 - Crie a base de treino e de teste.
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.3, random_state= 28)
```

```python
modelo = RandomForestClassifier()

modelo.fit(xtrain,ytrain)
```

#### Avaliando o classificador

```python
#5 - Avalie o classificador
print(f'{round(modelo.score(xtest,ytest),3) * 100}%')
```
