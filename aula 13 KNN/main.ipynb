### K vizinhos próximos

- Medida de distância (Euclidiana)
- K ímpar
- Voto do majoritário
- Média (regressão)
- Não é linear

As árvores também são modelos não linears, trabalhando ou com entropia ou ganho de informação, assim como o K means que utiliza a distância entre os vizinhos (não leva em consideração a propoção da distância entre os dados) [As distâncias não sâo uniformes




### Vantagens

- Simples implementação
- Resistente a ruidos (o algoritmo olha apenas os dados próximos do que deseja classificar)
- Poucos hiperparãmetros (apenas o número de vizinhos)

```python
# Importando bibliotecas necessárias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Gerando um dataset fictício
np.random.seed(0)
n_samples = 100
X = np.random.rand(n_samples, 3)  # 3 variáveis independentes
y = np.random.randint(0, 2, n_samples)  # Variável alvo binária (0 ou 1)

# Convertendo para DataFrame
df = pd.DataFrame(X, columns=['X', 'Y', 'Z'])
df['target'] = y

# Criando uma visualização 3D
fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot(111, projection='3d')

# Plotando os pontos do dataset
ax.scatter(df['X'], df['Y'], df['Z'], c=df['target'], marker='o', cmap='coolwarm')

# Definindo os rótulos dos eixos
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Título do gráfico
plt.title('Visualização 3D de um Dataset Fictício')

# Exibindo o gráfico
plt.show()
```
